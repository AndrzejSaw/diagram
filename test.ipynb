```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('stavka.csv')
# Размер в пикселях
width_px = 1280
height_px = 720

# Переводим пиксели в дюймы
width_in = width_px / 100.0
height_in = height_px / 100.0

# Устанавливаем размер фигуры
plt.figure(figsize=(width_in, height_in))
colors = ['green', 'orange', 'skyblue']
bars = plt.bar(data['Название фирмы'], data['Ставка (euro)'], color=colors)
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=1.5)
plt.xlabel('Название фирмы')
plt.ylabel('Ставка за день')
plt.title('Ставка за день по фирме X')
plt.xticks(rotation=15)  # Повернуть метки даты, если они слишком густые
for bar in bars:
    yval = bar.get_height()  # Получаем высоту столбца
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')
plt.gca().set_facecolor('#ebede1')  # Изменение фона графика
plt.gcf().set_facecolor('#e0cce6')  # Изменение фона вокруг графика

plt.show()

```
