```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('stavka.csv')
#   
width_px = 1280
height_px = 720

#    
width_in = width_px / 100.0
height_in = height_px / 100.0

#   
plt.figure(figsize=(width_in, height_in))
colors = ['green', 'orange', 'skyblue']
bars = plt.bar(data[' '], data[' (euro)'], color=colors)
plt.grid(True, which='both', axis='y', linestyle='--', linewidth=1.5)
plt.xlabel(' ')
plt.ylabel('  ')
plt.title('     X')
plt.xticks(rotation=15)  #   ,    
for bar in bars:
    yval = bar.get_height()  #   
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')
plt.gca().set_facecolor('#ebede1')  #   
plt.gcf().set_facecolor('#e0cce6')  #    

plt.show()

```
